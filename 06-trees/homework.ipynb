```python
import pandas as pd
```

```python
df = pd.read_csv('jamb_exam_results.csv', keep_default_na=False, na_values=['', 'NaN', 'null'])
```

```python
df
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
del df['student_id']
```

```python
df.fillna(0)
```

```python
df.head()
```

```python
df['parent_education_level'] = df['parent_education_level'].fillna(0)
```

```python
from sklearn.model_selection import train_test_split

df_full_train, df_test = train_test_split(df, test_size=0.2, random_state=1)
df_train, df_val = train_test_split(df_full_train, test_size=0.25, random_state=1)
```

```python
df_train = df_train.reset_index(drop=True)
df_val = df_val.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)
```

```python
y_train = df_train.jamb_score.values
y_val = df_val.jamb_score.values
y_test = df_test.jamb_score.values
```

```python
del df_train['jamb_score']
del df_val['jamb_score']
del df_test['jamb_score']
```

```python
from sklearn.feature_extraction import DictVectorizer
```

```python
dv = DictVectorizer(sparse=True)

train_dicts = df_train.to_dict(orient='records')
X_train = dv.fit_transform(train_dicts)

val_dicts = df_val.to_dict(orient='records')
X_val = dv.transform(val_dicts)
```

### Question 1

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt = DecisionTreeRegressor(max_depth=1)
dt.fit(X_train, y_train)
```

```python
from sklearn.tree import export_text
```

```python
print(export_text(dt, feature_names=dv.get_feature_names_out()))
```

### Question 2

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import root_mean_squared_error
```

```python
rf = RandomForestRegressor(n_estimators=10, random_state=1, n_jobs=-1)
rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_val)
```

```python
root_mean_squared_error(y_val, y_pred)
```

### Question 3

```python
scores = []

for n in range(10, 201, 10):
    rf = RandomForestRegressor(n_estimators=n, random_state=1)
    rf.fit(X_train, y_train)

    y_pred = rf.predict(X_val)
    rmse = root_mean_squared_error(y_val, y_pred)

    scores.append((n, rmse))
```

```python
df_scores = pd.DataFrame(scores, columns=['n_estimators', 'rmse'])
```

```python
df_scores
```

```python
from matplotlib import pyplot as plt 
```

```python
plt.plot(df_scores.n_estimators, df_scores.rmse)
```

### Question 4

```python
scores = []
mean_scores = []

for d in [10, 15, 20, 25]:
    for n in range(10, 201, 10):
        rf = RandomForestRegressor(n_estimators=n, max_depth=d, random_state=1)
        rf.fit(X_train, y_train)

        y_pred = rf.predict(X_val)
        rmse = root_mean_squared_error(y_val, y_pred)

        scores.append(rmse)

    mean = sum(scores) / len(scores)
    mean_scores.append((d, mean))

    scores = []
```

```python
import xgboost as xgb
```

```python
features = list(dv.get_feature_names_out())
dtrain = xgb.DMatrix(X_train, label=y_train, feature_names=features)
dval = xgb.DMatrix(X_val, label=y_val, feature_names=features)
```

```python
watchlist = [(dtrain, 'train'), (dval, 'val')]
```

```python
scores = {}
```

```python
%%capture output

xgb_params = {
    'eta': 0.1, 
    'max_depth': 6,
    'min_child_weight': 1,
    
    'objective': 'reg:squarederror',
 
    'nthread': 8,
    'seed': 1,
    'verbosity': 1,
}

model = xgb.train(xgb_params, dtrain, evals=watchlist, verbose_eval=5, num_boost_round=100)
```
